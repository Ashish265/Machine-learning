```python
listcol = ["Actual Duration","Actual Effort","Adjusted Function point","First estimate"]
```

### Distribution plots

```python
for col in listcol:
    plt.figure(figsize=(12,8))
    sns.distplot(df[col])
    plt.grid()
    plt.show()
```

### Box plots

```python
for col in listcol:
    plt.figure(figsize=(12,8))
    sns.boxplot(df[col])
    plt.grid()
    plt.show()
```
